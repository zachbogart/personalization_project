# Make Smaller IMDB csv

### Setup

```python
import pandas as pd
```

```python
movielens_data_links_path = '../working/ml-latest-small/links.csv'
imdb_data_path = '../working/title.basics.tsv'

csv_path = "imdb_side_info.csv"
```

```python
movies = pd.read_csv(movielens_data_links_path)

# movies.head()

movies = list(movies['imdbId'].astype(int))
```

### Read in Data (Just the good stuff)

```python
import csv
from datetime import datetime
startTime = datetime.now()

# read in data
imdb_reader = csv.reader(open(imdb_data_path), delimiter="\t")

results = []
columns = []

line = 0

# for every line
for row in imdb_reader:
    # header
    if line == 0:
        columns.extend(row) # add on headers
        line += 1
        continue
    else:
        imdbId_from_reader = int(row[0][2:]) # format for tconst
        if imdbId_from_reader in movies:
#             print(imdbId_from_reader)
            results.append(row)

    line += 1
    
resultsDF = pd.DataFrame(results)

print("Done")
print(len(results))
print(datetime.now() - startTime)
```

```python
resultsDF.columns = columns
resultsDF
```

### Split Genres 

```python
# foo = resultsDF.copy() # copy for testing
foo = resultsDF # now link directly
foo.head()
```

```python
foo.genres = foo['genres'].str.split(pat=",")
```

```python
# make list of all genres
all_genres = set()

for line in foo.genres:
    for item in line:
        all_genres.add(item)
        
all_genres_list = list(all_genres)
```

```python
all_genres_list
```

```python
for genre in all_genres_list:
    foo[genre] = 0
```

```python
foo
```

```python
for index, row in foo.iterrows():
    for item in row['genres']:
        foo.loc[index, item] = 1
```

```python
foo
```

### Export to CSV

```python
resultsDF.to_csv(csv_path)
print("Made CSV")
```
